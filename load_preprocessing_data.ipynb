## Missing values processing

```python
def check_missing(df, name):
    print(f"Missing values in {name}:")
    print(df.isnull().sum())
    print("-" * 40)

check_missing(df_construction, "df_construction")
check_missing(df_gdp, "df_gdp")
check_missing(df_industrial, "df_industrial")
check_missing(df_crane, "df_crane")
check_missing(df_excavator, "df_excavator")
check_missing(df_roller, "df_roller")
```
