## Missing values processing

```python
def check_missing(df, name):
    print(f"Missing values in {name}:")
    print(df.isnull().sum())
    print("-" * 40)

check_missing(df_construction, "df_construction")
check_missing(df_gdp, "df_gdp")
check_missing(df_industrial, "df_industrial")
check_missing(df_crane, "df_crane")
check_missing(df_excavator, "df_excavator")
check_missing(df_roller, "df_roller")
```

## Build of X and Y

```python
arr_crane = df_crane.iloc[:, 1:].to_numpy().flatten()
arr_excavator = df_excavator.iloc[:, 1:].to_numpy().flatten()
arr_roller = df_roller.iloc[:, 1:].to_numpy().flatten()
X_arr_2d = np.array([arr_crane, arr_excavator, arr_roller])
X_arr = X_arr_2d.T
arr_gdp = df_gdp.iloc[:, 1:-1].to_numpy().flatten()
Y_arr = arr_gdp

print(f"X_arr:{X_arr.shape}")
print(f"Y_arr:{Y_arr.shape}")
```

## Split into test and train

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

X_train, X_test, Y_train, Y_test = train_test_split(X_arr, Y_arr, test_size=0.2, random_state=42)
```

## Train and evaluate of the model

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, r2_score


# Define the models to compare, including MLP Regressor
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree Regression": DecisionTreeRegressor(random_state=42),
    "Random Forest Regression": RandomForestRegressor(random_state=42),
    "Support Vector Regression": SVR(),
    "MLP Regressor": MLPRegressor(random_state=42, max_iter=2000)
}

results = {}

# Train each model, predict on the test set, and calculate performance metrics
for name, model in models.items():
    model.fit(X_train, Y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(Y_test, y_pred)
    r2 = r2_score(Y_test, y_pred)
    results[name] = {"MSE": mse, "R²": r2}

# Print out the performance comparison for each model
print("Performance comparison of models on the test set:")
for name, metrics in results.items():
    print(f"{name}: MSE = {metrics['MSE']:.4f}, R² = {metrics['R²']:.4f}")
```

```python
def plot_gdp():
    df = df_construction.T
    df.columns = df.iloc[0]
    df = df[1:]
    df = df.iloc[::-1]
    df.rename_axis('Date')

    plt.plot(df.iloc[:, 1:], alpha=.5)
    plt.xlabel('Date')
    plt.ylabel('GDP')
    plt.title('GDP of provinces over time')
    _ = plt.xticks(rotation='vertical')
    plt.show()
```
