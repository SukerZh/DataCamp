## load data

```python
# read csv file
df_construction = pd.read_csv("ConstructionTotalValueOutput.csv",encoding='gbk', engine='python')
df_gdp = pd.read_csv("GrossDomesticProduct.csv",encoding='gbk', engine='python')
df_industrial = pd.read_csv("IndustrialVA_YonY.csv",encoding='gbk', engine='python')
df_crane = pd.read_csv("crane.csv",encoding='gbk', engine='python')
df_excavator = pd.read_csv("excavator.csv",encoding='gbk', engine='python')
df_roller = pd.read_csv("roller.csv",encoding='gbk', engine='python')

print("DataFrames loaded successfully!")
```

## Missing values processing

```python
def check_missing(df, name):
    print(f"Missing values in {name}:")
    print(df.isnull().sum())
    print("-" * 40)

check_missing(df_construction, "df_construction")
check_missing(df_gdp, "df_gdp")
check_missing(df_industrial, "df_industrial")
check_missing(df_crane, "df_crane")
check_missing(df_excavator, "df_excavator")
check_missing(df_roller, "df_roller")
```
